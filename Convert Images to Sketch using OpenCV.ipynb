```python
rev_img=img[:,:,::-1]
```

```python
plt.imshow(rev_img)
plt.axis(False)
```

```python
def show_image(img_arr):
    rgb_arr=cv2.cvtColor(img_arr,cv2.COLOR_BGR2RGB)
    plt.imshow(rgb_arr)
    plt.axis(False)
```

```python
grey_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
show_image(grey_img)
```

```python
show_image(255-grey_img)
inverted_image=255-grey_img
```

```python
blur_image=cv2.GaussianBlur(inverted_image,(111,111),0)
```

```python
show_image(blur_image)
```

```python
sketch=cv2.divide(blur_image,inverted_image,scale=180.0)
```

```python
show_image(sketch)
```
